````python
INPUT_SHAPE = (84, 84)
WINDOW_LENGTH = 4

env_name = 'SpaceInvaders-v0'
env = gym.make(env_name)

np.random.seed(123)
env.seed(123)
nb_actions = env.action_space.n

print("Numero de acciones disponibles: " + str(nb_actions) + " -> " + str(env.unwrapped.get_action_meanings()))
print("Formato de las observaciones: " + str(env.observation_space))
````

````python
class AtariProcessor(Processor):
    def process_observation(self, observation):
        assert observation.ndim == 3  # (height, width, channel)
        # removemos la parte superior (score) y inferior (vida del jugador)
        cropped = observation[20:200, 8:152]
        img = Image.fromarray(cropped)
        img = img.resize(INPUT_SHAPE, Image.LANCZOS).convert('L')
        processed_observation = np.array(img)
        assert processed_observation.shape == INPUT_SHAPE
        return processed_observation.astype('uint8')

    def process_state_batch(self, batch):
        processed_batch = batch.astype('float32') / 255.
        return processed_batch

    def process_reward(self, reward):
        return np.clip(reward, -1., 1.)
````

````python
# modelo mas complejo - usar este
input_shape = (WINDOW_LENGTH,) + INPUT_SHAPE
model = Sequential()
print(K.image_data_format())
print(input_shape)

if K.image_data_format() == 'channels_last':
    # (width, height, channels)
    model.add(Permute((2, 3, 1), input_shape=input_shape))
elif K.image_data_format() == 'channels_first':
    # (channels, width, height)
    model.add(Permute((1, 2, 3), input_shape=input_shape))
else:
    raise RuntimeError('Unknown image_dim_ordering.')

# Capas convolucionales mejoradas con inicialización HeNormal
model.add(Convolution2D(32, (8, 8), strides=(4, 4), 
                       activation='relu', kernel_initializer=HeNormal()))
 # model.add(BatchNormalization())

model.add(Convolution2D(64, (4, 4), strides=(2, 2), 
                       activation='relu', kernel_initializer=HeNormal()))
 # model.add(BatchNormalization())

model.add(Convolution2D(64, (3, 3), strides=(1, 1), 
                       activation='relu', kernel_initializer=HeNormal()))
 # model.add(BatchNormalization())

model.add(Flatten())

# Capas densas con dropout y batch normalization
model.add(Dense(512, activation='relu', kernel_initializer=HeNormal()))
 # model.add(BatchNormalization())
model.add(Dropout(0.3))

model.add(Dense(256, activation='relu', kernel_initializer=HeNormal()))
 # model.add(BatchNormalization())
model.add(Dropout(0.2))

# Capa de salida
model.add(Dense(nb_actions, activation='linear', kernel_initializer=HeNormal()))

print(model.summary())
````

2. Implementación de la solución DQN

````python
import numpy as np
import random
from collections import deque
from rl.memory import Memory
import warnings

class SumTree:
    """
    Estructura de datos Sum Tree para muestreo eficiente basado en prioridades.
    """
    def __init__(self, capacity):
        self.capacity = capacity
        self.tree = np.zeros(2 * capacity - 1)
        self.data = np.zeros(capacity, dtype=object)
        self.n_entries = 0
        self.pending_idx = set()

    def _propagate(self, idx, change):
        """Propaga el cambio hacia arriba en el árbol"""
        parent = (idx - 1) // 2
        self.tree[parent] += change
        if parent != 0:
            self._propagate(parent, change)

    def _retrieve(self, idx, s):
        """Recupera el índice de la hoja correspondiente al valor s"""
        left = 2 * idx + 1
        right = left + 1

        if left >= len(self.tree):
            return idx

        if s <= self.tree[left]:
            return self._retrieve(left, s)
        else:
            return self._retrieve(right, s - self.tree[left])

    def total(self):
        """Retorna la suma total (raíz del árbol)"""
        return self.tree[0]

    def add(self, p, data):
        """Añade una nueva experiencia con prioridad p"""
        idx = self.n_entries + self.capacity - 1
        self.pending_idx.add(self.n_entries)
        
        self.data[self.n_entries] = data
        self.update(idx, p)

        self.n_entries += 1
        if self.n_entries >= self.capacity:
            self.n_entries = 0

    def update(self, idx, p):
        """Actualiza la prioridad en el índice idx"""
        change = p - self.tree[idx]
        self.tree[idx] = p
        self._propagate(idx, change)

    def get(self, s):
        """Obtiene una muestra basada en el valor s"""
        idx = self._retrieve(0, s)
        dataIdx = idx - self.capacity + 1
        self.pending_idx.discard(dataIdx)
        return (idx, self.tree[idx], self.data[dataIdx])


class PrioritizedExperienceReplay(Memory):
    """
    Implementación de Prioritized Experience Replay compatible con Keras-RL
    """
    def __init__(self, limit, alpha=0.6, beta_start=0.4, beta_frames=100000, 
                 epsilon=1e-6, window_length=1, ignore_episode_boundaries=False):
        super(PrioritizedExperienceReplay, self).__init__(window_length, ignore_episode_boundaries)
        
        self.limit = limit
        self.alpha = alpha  # Cuánto usar priorización (0 = uniforme, 1 = completamente priorizado)
        self.beta_start = beta_start  # Corrección de importancia inicial
        self.beta_frames = beta_frames  # Frames para llegar a beta=1.0
        self.epsilon = epsilon  # Pequeña cantidad para evitar prioridades cero
        self.frame_idx = 0  # Contador de frames para beta annealing
        
        # Estructuras de datos
        self.tree = SumTree(limit)
        self.experiences = deque(maxlen=limit)
        self.max_priority = 1.0
        
        # Para compatibilidad con keras-rl
        self.recent_observations = deque(maxlen=window_length)
        self.recent_actions = deque(maxlen=window_length)
        self.recent_rewards = deque(maxlen=window_length)
        self.recent_terminals = deque(maxlen=window_length)
        
    def beta_by_frame(self, frame_idx):
        """Annealing de beta desde beta_start hasta 1.0"""
        return min(1.0, self.beta_start + frame_idx * (1.0 - self.beta_start) / self.beta_frames)
    
    def append(self, observation, action, reward, terminal, training=True):
        """Añade una experiencia a la memoria"""
        #if reward != 0:
        #    print(f"append: {reward}")
        super(PrioritizedExperienceReplay, self).append(observation, action, reward, terminal, training)
        
        if training:
            # Solo crear experiencia si tenemos al menos una observación previa
            if len(self.recent_observations) >= 2:
                # Crear la experiencia
                experience = {
                    'observation0': self.recent_observations[-2],  # Observación anterior
                    'action': action, 
                    'reward': reward,
                    'observation1': observation,  # Observación actual
                    'terminal1': terminal,
                }
                
                # Añadir con prioridad máxima (para nuevas experiencias)
                priority = self.max_priority ** self.alpha
                self.tree.add(priority, experience)
                self.experiences.append(experience)
            elif len(self.recent_observations) == 1:
                # Para la primera experiencia, usar la misma observación
                experience = {
                    'observation0': self.recent_observations[-1],
                    'action': action,
                    'reward': reward, 
                    'observation1': observation,
                    'terminal1': terminal,
                }
                
                priority = self.max_priority ** self.alpha
                self.tree.add(priority, experience)
                self.experiences.append(experience)
    
    def sample(self, batch_size, batch_idxs=None):
        """Muestrea un batch de experiencias basado en prioridades"""
        print(f"sample: {batch_size}")
        if batch_idxs is not None:
            warnings.warn('PrioritizedExperienceReplay ignora batch_idxs')
        
        if len(self.experiences) == 0:
            raise ValueError('No hay experiencias para muestrear')
        
        # Inicializar listas para el batch
        batch = []
        idxs = []
        priorities = []
        
        # Muestreo basado en prioridades
        beta = self.beta_by_frame(self.frame_idx)
        self.frame_idx += 1
        
        if self.tree.total() == 0 or len(self.experiences) < batch_size:
            # Muestreo uniforme si el árbol está vacío o no hay suficientes experiencias
            for i in range(batch_size):
                idx = random.randint(0, len(self.experiences) - 1)
                batch.append(self.experiences[idx])
                idxs.append(idx)
                priorities.append(self.max_priority)
        else:
            # Muestreo normal basado en prioridades
            segment = self.tree.total() / batch_size
            
            for i in range(batch_size):
                a = segment * i
                b = segment * (i + 1)
                s = random.uniform(a, b)
                
                try:
                    (idx, p, data) = self.tree.get(s)
                    if data is not None:
                        batch.append(data)
                        idxs.append(idx)
                        priorities.append(p)
                    else:
                        # Fallback si no se puede obtener data
                        fallback_idx = random.randint(0, len(self.experiences) - 1)
                        batch.append(self.experiences[fallback_idx])
                        idxs.append(fallback_idx)
                        priorities.append(self.max_priority)
                except:
                    # Fallback en caso de error
                    fallback_idx = random.randint(0, len(self.experiences) - 1)
                    batch.append(self.experiences[fallback_idx])
                    idxs.append(fallback_idx)
                    priorities.append(self.max_priority)
        
        # Calcular pesos de importancia
        if self.tree.total() > 0:
            sampling_probabilities = np.array(priorities) / self.tree.total()
            is_weight = np.power(len(self.experiences) * sampling_probabilities, -beta)
            is_weight /= is_weight.max()  # Normalizar
        else:
            is_weight = np.ones(batch_size)
        
        # Formatear batch para keras-rl con dimensiones correctas
        experiences = []
        for exp in batch:
            # Asegurar que las observaciones tengan la dimensión correcta
            state0 = exp['observation0']
            state1 = exp['observation1']
            
            # Si las observaciones no tienen la dimensión del window_length, agregarla
            if len(state0.shape) == 2:  # (height, width) -> (window_length, height, width)
                state0 = np.expand_dims(state0, axis=0)
            if len(state1.shape) == 2:  # (height, width) -> (window_length, height, width)
                state1 = np.expand_dims(state1, axis=0)
            
            # Si window_length > 1, repetir la observación para llenar la ventana
            if self.window_length > 1:
                if state0.shape[0] < self.window_length:
                    state0 = np.repeat(state0, self.window_length, axis=0)[:self.window_length]
                if state1.shape[0] < self.window_length:
                    state1 = np.repeat(state1, self.window_length, axis=0)[:self.window_length]
            
            # Crear un objeto Experience compatible con keras-rl
            experience_obj = type('Experience', (), {
                'state0': state0,
                'action': exp['action'],
                'reward': exp['reward'],
                'state1': state1,
                'terminal1': exp['terminal1']
            })()
            experiences.append(experience_obj)
        
        # Almacenar información adicional para PER
        self._last_batch_idxs = idxs
        self._last_batch_priorities = priorities
        self._last_batch_weights = is_weight

        print(f"Se seleccionaron estas experiencias del buffer: {experiences}")
        return experiences
    
    def update_priorities(self, batch_idxs, batch_priorities):
        """Actualiza las prioridades de las experiencias muestreadas"""
        print(f"update_priorities: {batch_priorities}")
        for idx, priority in zip(batch_idxs, batch_priorities):
            priority = abs(priority) + self.epsilon  # Evitar prioridades cero
            self.max_priority = max(self.max_priority, priority)
            
            clipped_priority = priority ** self.alpha
            self.tree.update(idx, clipped_priority)
    
    def get_last_batch_info(self):
        """Retorna información del último batch muestreado para actualizar prioridades"""
        return {
            'idxs': getattr(self, '_last_batch_idxs', []),
            'priorities': getattr(self, '_last_batch_priorities', []),
            'weights': getattr(self, '_last_batch_weights', [])
        }
    
    def __len__(self):
        """Retorna el número de experiencias almacenadas"""
        return len(self.experiences)
    
    @property
    def nb_entries(self):
        """Propiedad para compatibilidad con keras-rl"""
        return len(self.experiences)
````

````python
import numpy as np
from collections import deque
from rl.memory import Memory  # asegurate que viene de keras-rl2

class ElitistMemory(Memory):

    def __init__(self, limit, window_length, ignore_episode_boundaries=False):
        super(ElitistMemory, self).__init__(window_length, ignore_episode_boundaries)
        self.limit = limit
        self.window_length = window_length
        self.data = []

        # Trackeo del episodio actual
        self._episode_reward = 0.0
        self._episode_experiences = []

        # Requerido por keras-rl (aunque Memory ya lo define, lo explicitamos)
        self.recent_observations = deque(maxlen=window_length)
        self.recent_actions = deque(maxlen=window_length)
        self.recent_rewards = deque(maxlen=window_length)
        self.recent_terminals = deque(maxlen=window_length)

    def append(self, observation, action, reward, terminal, training=True):
        super().append(observation, action, reward, terminal, training)

        if not training:
            return

        self._episode_reward += reward

        # Construir experiencia con lo que haya disponible
        if all(len(x) >= 2 for x in [self.recent_observations, self.recent_actions, self.recent_rewards, self.recent_terminals]):
            obs0 = self.recent_observations[-2]
            obs1 = self.recent_observations[-1]
            act = self.recent_actions[-2]
            rew = self.recent_rewards[-2]
            term = self.recent_terminals[-2]
        elif len(self.recent_observations) >= 1:
            obs0 = self.recent_observations[-1]
            obs1 = observation
            act = action
            rew = reward
            term = terminal
        else:
            return

        experience = {
            'observation0': obs0,
            'action': act,
            'reward': rew,
            'observation1': obs1,
            'terminal1': term,
            'episode_reward': None  # se asigna al final del episodio
        }

        self._episode_experiences.append(experience)

        if terminal:
            # Guardar las experiencias del episodio con su reward total
            for exp in self._episode_experiences:
                exp['episode_reward'] = self._episode_reward
                self._insert_experience(exp)

            print(f"/n[MEMORY] Episodio terminado con reward total: {self._episode_reward:.2f}")
            self._episode_reward = 0.0
            self._episode_experiences.clear()

    def _insert_experience(self, experience):
        if len(self.data) < self.limit:
            self.data.append(experience)
            print(f"/n[MEMORY] Agregada experiencia (ep_reward={experience['episode_reward']:.2f})")
        else:
            rewards = [e['episode_reward'] for e in self.data]
            worst_idx = np.argmin(rewards)
            worst_reward = rewards[worst_idx]

            if experience['episode_reward'] > worst_reward:
                print(f"/n[MEMORY] Reemplazada experiencia (old={worst_reward:.2f}) -> nueva={experience['episode_reward']:.2f}")
                self.data[worst_idx] = experience
            else:
                print(f"/n[MEMORY] Descartada experiencia (ep_reward={experience['episode_reward']:.2f} < {worst_reward:.2f})")

    def sample(self, batch_size, batch_idxs=None):
        if len(self.data) < batch_size:
            raise ValueError("No hay suficientes experiencias para samplear.")
        indices = np.random.choice(len(self.data), batch_size, replace=False)
        batch = [self.data[i] for i in indices]

        # Adaptar al formato que espera keras-rl
        experiences = []
        for exp in batch:
            experience_obj = type('Experience', (), {
                'state0': exp['observation0'],
                'action': exp['action'],
                'reward': exp['reward'],
                'state1': exp['observation1'],
                'terminal1': exp['terminal1']
            })()
            experiences.append(experience_obj)

        return experiences

    def __len__(self):
        return len(self.data)

    @property
    def nb_entries(self):
        return len(self.data)
````

````python
from rl.agents.dqn import DQNAgent

class DQNPERAgent(DQNAgent):
    def backward(self, reward, terminal):
        metrics = super().backward(reward, terminal)
        print(f"DQNPERAgent.backward: {reward}")
        if hasattr(self.memory, "update_priorities") and hasattr(self.memory, "get_last_batch_info"):
            batch_info = self.memory.get_last_batch_info()
            if batch_info['idxs']:
                td_errors = self.delta  # Esto es un atributo interno de keras-rl que contiene el error TD
                self.memory.update_priorities(batch_info['idxs'], td_errors)

        return metrics
````

````python
from rl.callbacks import Callback

class HyperparameterMonitorCallback(Callback):
    """Callback para monitorear el estado de la memoria de los actuales valores de los hiperparametros"""
    
    def __init__(self, interval=5000):
        self.interval = interval
        self.steps_since_log = 0
        
    def on_step_end(self, step, logs=None):
        self.steps_since_log += 1
        if self.steps_since_log >= self.interval:
            self.steps_since_log = 0
            
            print(f"\n>>> Memoria e Hiperparametros:")
            
            # Estado de la memoria/buffer
            memory = self.model.memory
            print(f"    [MEMORY] Memory Size: {len(memory)}")
            print(f"    [MEMORY] Elite Fraction: {memory.elite_fraction}")
            
            # Valores de los hiperparametros
            print(f"    [TRAIN] Batch Size: {dqn.batch_size}")
            print(f"    [TRAIN] Step actual: {self.model.step}")
            
            policy = getattr(self.model, 'policy', None)
            if isinstance(policy, LinearAnnealedPolicy):
                eps = policy.get_current_value()
                print(f"    [POLICY] epsilon actual: {eps:.4f}")
            optimizer = self.model.model.optimizer
            lr_schedule = optimizer.learning_rate

            if isinstance(lr_schedule, ExponentialDecay):
                # Usamos el número de paso actual para evaluar el schedule
                current_lr = float(lr_schedule(self.model.step).numpy())
            else:
                # Puede ser una variable o valor fijo
                current_lr = float(K.get_value(lr_schedule))
            print(f"    [LR] Learning rate actual: {current_lr:.6f}")
            

    def on_episode_end(self, episode, logs=None):
        """Mostrar stats al final del episodio"""
        pass
````

````python
from rl.callbacks import Callback

class BatchSizeSchedulerCallback(Callback):
    def __init__(self, total_steps, initial_bs=32, schedule=None):
        self.total_steps = total_steps
        self.initial_bs = initial_bs
        self.schedule = schedule if schedule else [(1/3, 2), (2/3, 2)]
        self.applied = [False] * len(self.schedule)

    def on_train_begin(self, logs=None):
        self.model.batch_size = self.initial_bs
        print(f"[BATCH] Entrenamiento iniciado con batch_size={self.initial_bs}")

    def on_step_end(self, step, logs=None):
        current_step = self.model.step  # paso global del agente

        for i, (threshold_frac, multiplier) in enumerate(self.schedule):
            if not self.applied[i] and current_step >= int(self.total_steps * threshold_frac):
                new_bs = self.model.batch_size * multiplier
                self.model.batch_size = new_bs
                self.applied[i] = True
                print(f"[BATCH] Step {current_step}: batch_size actualizado a {new_bs}")
````

````python
from rl.callbacks import Callback
import numpy as np

class SaveBestRewardCallback(Callback):
    def __init__(self, env,
                 filename='best_model.h5f',
                 test_episodes=5,
                 test_interval=1000):
        self.env = env
        self.filename = filename
        self.test_episodes = test_episodes
        self.test_interval = test_interval
        self.best_reward = -np.inf
        self.steps_since_last_eval = 0

    def on_step_end(self, step, logs=None):
        self.steps_since_last_eval += 1

        if self.steps_since_last_eval >= self.test_interval:
            self.steps_since_last_eval = 0

            # --- congelar contador ---
            saved_step = self.model.step

            # --- evaluación ---
            history = self.model.test(
                self.env,
                nb_episodes=self.test_episodes,
                visualize=False,
                verbose=0
            )
            
            avg_reward = np.mean(history.history['episode_reward'])
            max_reward = np.max(history.history['episode_reward'])
            std_reward = np.std(history.history['episode_reward'])

            # --- restaurar contador ---
            self.model.step = saved_step
            self.model.training = True

            print(f"\n>>> Step {saved_step} -> Avg={avg_reward:.2f}, Max={max_reward:.2f} -> {history.history['episode_reward']} -> Mejor Actual: {self.best_reward:.2f}")

            if avg_reward > self.best_reward:
                self.best_reward = avg_reward
                self.model.save_weights(self.filename, overwrite=True)
                print(f">>> Nuevo Máximo: {avg_reward:.2f} -> Guardado en {self.filename}")
````

````python
from rl.callbacks import Callback
import numpy as np

class AdaptiveEliteFractionCallback(Callback):
    def __init__(self, 
                 env,
                 monitor_interval=1000,
                 improve_threshold=1.0,
                 decay_threshold=0.1,
                 decay_factor=0.7,
                 grow_factor=0.7,
                 min_elite=0.1,
                 max_elite=0.9):
        self.env = env
        self.monitor_interval = monitor_interval
        self.improve_threshold = improve_threshold
        self.decay_threshold = decay_threshold
        self.decay_factor = decay_factor
        self.grow_factor = grow_factor
        self.min_elite = min_elite
        self.max_elite = max_elite
        self.best_avg_reward = -np.inf
        self.steps_since_last_eval = 0

    def on_step_end(self, step, logs={}):
        self.steps_since_last_eval += 1

        if self.steps_since_last_eval >= self.monitor_interval:
            self.steps_since_last_eval = 0
            saved_step = self.model.step

            # Evaluar desempeño actual
            history = self.model.test(
                self.env,
                nb_episodes=5,
                visualize=False,
                verbose=0
            )

            avg_reward = np.mean(history.history['episode_reward'])
            delta = avg_reward - self.best_avg_reward

            mem = self.model.memory
            current_elite = getattr(mem, 'elite_fraction', None)
            if current_elite is None:
                print(f"\n[AdaptiveElite] Error: la memoria no tiene 'elite_fraction'")
                return

            print(f"\n[AdaptiveElite] Step {saved_step} -> Avg={avg_reward:.2f}, Best={self.best_avg_reward:.2f}, Δ={delta:.2f}, elite_fraction={current_elite:.2f}")

            # Ajustar elite_fraction dinámicamente
            if delta > self.improve_threshold:
                new_elite = min(current_elite * self.grow_factor, self.max_elite)
                self.best_avg_reward = avg_reward
                print(f"\n[AdaptiveElite] Mejora detectada → Aumento elite_fraction a {new_elite:.2f}")
                mem.elite_fraction = new_elite

            elif delta < self.decay_threshold:
                new_elite = max(current_elite * self.decay_factor, self.min_elite)
                print(f"\n[AdaptiveElite] Estancamiento detectado → Reduzco elite_fraction a {new_elite:.2f}")
                mem.elite_fraction = new_elite
````

````python
import numpy as np
import copy
import random
from collections import deque
from rl.memory import sample_batch_indexes, zeroed_observation

class ElitistSequentialMemory(SequentialMemory):
    def __init__(self, limit, window_length=1, elite_fraction=0.5, batch_size=128, reward_threshold=0.0, **kwargs):
        super().__init__(limit=limit, window_length=window_length, **kwargs)
        self.elite_fraction = elite_fraction
        self.batch_size = batch_size
        self.reward_threshold = reward_threshold
        self.limit = limit
        self.window_length = window_length

        self.elite_fraction = elite_fraction  # porcentaje del batch que será elite
        self.reward_threshold = reward_threshold  # criterio de recompensa mínima para ser elite
        
        self.actions = []
        self.rewards = []
        self.terminals = []
        self.observations = []

    def append(self, observation, action, reward, terminal, training=True):
        super().append(observation, action, reward, terminal, training=training)
        if not training:
            return

        if reward < 0:
            print(f"\n[MEMORY] NEGATIVA {reward:.2f}")

        if len(self.observations) < self.limit:
            self.observations.append(observation)
            self.actions.append(action)
            self.rewards.append(reward)
            self.terminals.append(terminal)
            # print(f"\n[MEMORY] Agregada experiencia nueva con reward {reward:.2f}")
        else:
            # Buscar el índice de la peor experiencia actual
            min_reward = min(self.rewards)
            if reward > min_reward:
                idx = self.rewards.index(min_reward)
                #if reward > 0:
                #    print(f"\n[MEMORY] Reemplazada exp (old: {min_reward:.2f}) -> nueva: {reward:.2f}")
                self.observations[idx] = observation
                self.actions[idx] = action
                self.rewards[idx] = reward
                self.terminals[idx] = terminal

    def sample(self, batch_size, batch_idxs=None):
        all_experiences = super().sample(batch_size * 4)
    
        # Definimos una recompensa mínima para considerar élite (podés ajustar)
        rewards = [e.reward for e in all_experiences]
        threshold = np.percentile(rewards, 100 * (1 - self.elite_fraction))  # por ejemplo, top 20%
    
        elite = [e for e in all_experiences if e.reward >= threshold]
        elite_ids = set(id(e) for e in elite)
    
        elite_count = min(len(elite), int(self.elite_fraction * batch_size))
        selected_elite = elite[:elite_count]
    
        remaining = batch_size - len(selected_elite)
        others = [e for e in all_experiences if id(e) not in elite_ids]
    
        if len(others) >= remaining:
            final_others = others[:remaining]
        else:
            # Si no hay suficientes "otros", completamos duplicando de los mejores
            needed = remaining - len(others)
            if len(others) > 0:
                final_others = others + random.choices(others, k=needed)
            else:
                final_others = random.choices(selected_elite, k=remaining)
        
        final_batch = selected_elite + final_others
        #print(f"\n[MEMORY] Se muestrea un batch de {batch_size} con {len(selected_elite)} exp de elite y {len(final_others)} restantes.")
        assert len(final_batch) == batch_size
        return final_batch

    @property
    def nb_entries(self):
        return len(self.observations)

    def get_config(self):
        config = super().get_config()
        config['limit'] = self.limit
        return config
    
    def __len__(self):
        return len(self.observations)
````

````python
from rl.callbacks import Callback
import numpy as np
import math

class AdaptiveEliteTuningAndSaveBestModelCallback(Callback):
    def __init__(self, 
                 env,
                 filename='best_model.h5f',
                 test_episodes=5,
                 warmup_steps=1000,
                 monitor_interval=1000,
                 improve_threshold=1.0,
                 decay_threshold=0.1,
                 decay_factor=0.7,
                 grow_factor=0.7,
                 min_elite=0.1,
                 max_elite=0.9):
        
        self.env = env
        self.filename = filename
        self.test_episodes = test_episodes
        self.warmup_steps = warmup_steps
        self.monitor_interval = monitor_interval
        self.improve_threshold = improve_threshold
        self.decay_threshold = decay_threshold
        self.decay_factor = decay_factor
        self.grow_factor = grow_factor
        self.min_elite = min_elite
        self.max_elite = max_elite
        self.best_avg_reward = -np.inf
        self.steps_since_last_eval = 0

    def on_step_end(self, step, logs=None):
        self.steps_since_last_eval += 1
        
        if self.steps_since_last_eval >= self.monitor_interval and self.steps_since_last_eval >= self.warmup_steps:
            self.steps_since_last_eval = 0
            saved_step = self.model.step

            # Evaluación
            history = self.model.test(
                self.env,
                nb_episodes=self.test_episodes,
                visualize=False,
                verbose=0
            )
            
            avg_reward = np.mean(history.history['episode_reward'])
            max_reward = np.max(history.history['episode_reward'])
            delta = avg_reward - self.best_avg_reward
            print(f"\n>>> {self.best_avg_reward}")

            # Restaurar contador del modelo (importante para evitar errores internos)
            self.model.step = saved_step
            self.model.training = True

            print(f"\n>>> Step {saved_step} -> Avg={avg_reward:.2f}, Max={max_reward:.2f} -> {history.history['episode_reward']} -> Mejor Actual: {self.best_avg_reward:.2f}")

            # Guardar pesos si mejora
            if avg_reward > self.best_avg_reward:
                self.best_avg_reward = avg_reward
                self.model.save_weights(self.filename, overwrite=True)
                print(f">>> Nuevo Máximo: {avg_reward:.2f} -> Guardado en {self.filename}")

            # Ajuste de elite_fraction si la memoria lo permite
            if not math.isinf(delta):
                mem = getattr(self.model, 'memory', None)
                current_elite = getattr(mem, 'elite_fraction', None)
    
                if current_elite is not None:
                    print(f"[AdaptiveElite] Δ={delta:.2f}, elite_fraction={current_elite:.2f}")
                    if delta > self.improve_threshold:
                        new_elite = min(current_elite * self.grow_factor, self.max_elite)
                        print(f"[AdaptiveElite] Mejora → Aumento elite_fraction a {new_elite:.2f}")
                        mem.elite_fraction = new_elite
                    elif delta < self.decay_threshold:
                        new_elite = max(current_elite * self.decay_factor, self.min_elite)
                        print(f"[AdaptiveElite] Estancamiento → Reduzco elite_fraction a {new_elite:.2f}")
                        mem.elite_fraction = new_elite
````
